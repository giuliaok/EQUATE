```python
def find_speakers(language):
    # Construct the Wikipedia URL for the language
    url = f"https://en.wikipedia.org/wiki/{language.replace(' ', '_')}"

    # Fetch the page content
    response = requests.get(url)
    if response.status_code != 200:
        return None  # If page not found, return None

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the infobox
    infobox = soup.find('table', class_='infobox')

    # Initialize speakers variable
    speakers = None

    # Find the row containing 'speakers'
    if infobox:
        rows = infobox.find_all('tr')
        for row in rows:
            if 'speakers' in row.text.lower():
                # Extract the number of speakers
                speakers = row.find('td').text.strip()
                break

    return speakers
```
